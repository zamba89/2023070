```python
import mysql.connector
import csv
import pandas as pd
```

```python
# Connect with MySQL
connection = mysql.connector.connect(
    host='localhost',
    user='root',
    password='Cbre2020',
    database='ca2'
)

# Create a cursor object to execute SQL queries:
cursor = connection.cursor()

# Execute a query to fetch the data from the 'ca2' table:
query = "SELECT * FROM ca2.year"
cursor.execute(query)

# Fetch all the data rows:
rows = cursor.fetchall()

# Define the file path where you want to save the CSV file:
csv_file_path = '/Users/AP139GE/MSc_CA2/Dataset/mysql.csv'

# Write the fetched data to the CSV file:

with open(csv_file_path, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow([i[0] for i in cursor.description])  # Write column headers
    writer.writerows(rows)  # Write data rows
    
# Close the cursor and the database connection
cursor.close()
connection.close()
```
